Podemos instanciar un cultivo iniciado con una única célula madre cancerosa:

```python
csc_culture = Culture(first_cell_is_stem=True)
csc_culture.simulate(1)
print(len(csc_culture.cells))
csc_culture.plot_culture_spheres()
```

Podemos modificar diversos parámetros como la probabilidad de que una célula madre de otra célula madre como resultado de su reproducción:

```python
another_csc_culture = Culture(
    first_cell_is_stem=True, prob_stem=0.5, cell_max_repro_attempts=1000
)
another_csc_culture.simulate(7)
print(len(another_csc_culture.cells))
another_csc_culture.plot_culture_spheres()
```

Podemos hacer una animación del proceso de reproducción:

```python
culture = Culture(
    first_cell_is_stem=True, prob_stem=0.5, cell_max_repro_attempts=1000
)
simulate_and_animate_growth(culture, num_steps=10, filename="growth.mp4")
```
