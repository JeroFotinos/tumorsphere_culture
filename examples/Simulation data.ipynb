```python
%matplotlib inline

from tumorsphere.cells import *
from tumorsphere.culture import *
```

```python
csc_culture = Culture(first_cell_is_stem=True)
```

```python
data = csc_culture.simulate_with_data(8)
```

```python
import matplotlib.pyplot as plt

# create a figure and axis objects
fig, ax = plt.subplots()

# plot each row of the array with custom labels and colors
ax.plot(data[0,:], label="total", color="blue")
ax.plot(data[1,:], label="active", color="green")
ax.plot(data[2,:], label="active CSCs", color="red")

# set the title and axis labels
ax.set_title("Rows of my array as functions of the columns")
ax.set_xlabel("Column index")
ax.set_ylabel("Value")

# create a legend and display the plot
ax.legend()
plt.show()
```
