```python
%matplotlib inline

from tumorsphere.cells import *
from tumorsphere.culture import *
```

```python
csc_culture = Culture(first_cell_is_stem=True)
```

```python
data = csc_culture.simulate_with_data(8)
```

```python
import matplotlib.pyplot as plt

# create a figure and axis objects
fig, ax = plt.subplots()

# plot each row of the array with custom labels and colors
ax.plot(data[0,:], label="Total", color="blue")
ax.plot(data[1,:], label="active", color="green")
ax.plot(data[2,:], label="active CSCs", color="red")

# set the title and axis labels
ax.set_title("Simulation of CSC seeded culture")
ax.set_xlabel("Time step")
ax.set_ylabel("Number of cells")

# create a legend and display the plot
ax.legend()
plt.show()
```

```python
import time

# start the timer
start_time = time.perf_counter()

# code
csc_culture = Culture(first_cell_is_stem=True)
data = csc_culture.simulate_with_data(8)

# stop the timer
end_time = time.perf_counter()

# calculate the elapsed time
elapsed_time = end_time - start_time

# print the elapsed time
print(f"Time taken: {elapsed_time:.6f} seconds")
```
